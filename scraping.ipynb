```python
from bs4 import BeautifulSoup
import requests
```

```python
def get(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    print(soup.prettify())
    return soup
soup=get("https://en.wikipedia.org/wiki/Linux")
```

```python
def title(soup):
    title = soup.select_one('title').text
    return title
title(soup)
```

```python
def dicto(soup):
    wiki = {}
    head = ''
    ps = []
    content = soup.select("div#mw-content-text>div>*")
    for c in content:
        if c.name == 'p':
            ps.append(c.text.strip())
        elif c.name == 'h2':
            wiki[head] = '\n'.join(ps).strip()
            head = c.get_text()
            ps = []
    return wiki

dicto(soup)                
```

```python
def lien(soup):
    A = [a.get('href', '') for a in soup.select("a")]
    A = [a for a in A if a.startswith('/wiki/')]
    return A
lien(soup)
```
